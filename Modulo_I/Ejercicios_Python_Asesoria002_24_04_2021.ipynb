```python
#construye un diccionario con una funcion
#que cuente el numero de letras y lo almacene
#'letra':'veces'
#ejemplo 'coco'
#{'c':'2','o':'2'}
word = 'brontosaurus'
d=dict()
for char in word: #recorrer la cadena de char
  if char not in d: #si el caracter no esta en el dicc
    d[char]=1 #
  else:
    d[char]=d[char]+1
print(d)
```

```python
#en la secuencia de ADN crea una diccionario que 
#cuente los nucleotidos y cree un diccionario
#cuantos aminoacidos Alanina (A) protein X
ADN = 'aatgattacgtagctagctagctagctagctagctagatagtagctagctcgt'
def contar_char(seq):
  dicc_ADN=dict()
  for nt in seq:
    if nt not in dicc_ADN:
      dicc_ADN[nt]=1
    else:
      dicc_ADN[nt]=dicc_ADN[nt]+1
  return dicc_ADN

contar_char(ADN)
```

```python
my_dicc ={"java":101, "python":102, "c":103}
#entrega los values y guardalos en una lista
#entrega los keys y guardalos en una lista
list_keys=list(my_dicc.keys())
list_values=list(my_dicc.values())
```

```python
#empleando list_values y list_keys reconstruye el diccionario
#en forma de tupla -->ZIP
#en forma de diccionario-->{ k:v for (k,v) in zip(list_keys, list_values)}
x=zip(list_keys,list_values)
print(list(x))#tuplas
y={k:v for (k,v) in zip(list_keys,list_values)} #list comprehension
print(y)
```
