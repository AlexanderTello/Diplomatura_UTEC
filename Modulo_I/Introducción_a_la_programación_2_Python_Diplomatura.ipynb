**B. La expresión While**. Las tareas repetitivas son comunes cuando las computadoras procesan información o realizan alguna función. Python ofrece funciones para ejecutar iteraciones. Una forma de iterar en Python es mediante la expresión `While`. `While` se ejecuta hasta que se cumpla la condición o enunciado.  

Por ejemplo, se tiene un programa que cuenta hacia atrás desde 5 y luego se imprime "Despegue". 

````python
n = 5
# Iteración
while n > 0: 
  print(n)
  n = n - 1

print("Despegue")
````

El código significa: `Mientras n sea mayor a cero, imprimir el valor de n y luego reducir en 1 el valor de n`. Finalmente se imprime la frase `"Despegue"`. Es decir,
```

  print(n)       # Imprimir el valor de n
  n = n - 1      # Reducir en 1 el valor de n
print("Acción")  # Imprimir la frase "Despegue"
```
