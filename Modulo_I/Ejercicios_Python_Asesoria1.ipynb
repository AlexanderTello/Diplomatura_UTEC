```python
#crear un diccionario con los especialistas
e={1000:{'nombre':'Sandra Loza','fecha_ing':'10-08-2017','dep':103},
   1001:{'nombre':'Fernando Dominguez','fecha_ing':'01-11-2018','dep':102},
   1002:{'nombre':'Francisco Velez','fecha_ing':'21-10-2019','dep':101},
   1003:{'nombre':'Laura Trelles','fecha_ing':'12-06-2016','dep':102}}
print('Base de datos de los medicos especialistas\n')
for clave,dic2 in e.items():
  print('Persona ID: ', clave)
  for key,value in dic2.items():
    print(key,':',value)
  print()

```
