>> **Realiza 8:** Usando las dos lista listas, crear un archivo que almacene el nombre, la secuencia en nucleótidos y su traducción en aminoácidos. Almacenarlos en el archivo `database3.txt`.



```
nombres = ['KF25', 'PL41', 'TR36' , 'VT44']

secuencias = ['CCATCGATCGTATCGATCTACACA',
              'CTAGCATTAATCGTGCATATCGTA',
              'CTAGCATACGATGCTAGCATTCTA',  
              'CATCGATCGATATTTGTATCTAGT' ]
```

El archivo database3.txt deberá contener lo siguiente:

image.png

````python
nombres = ['KF25', 'PL41', 'TR36' , 'VT44']

secuencias = ['CCATCGATCGTATCGATCTACACA',
              'CTAGCATTAATCGTGCATATCGTA',
              'CTAGCATACGATGCTAGCATTCTA',  
              'CATCGATCGATATTTGTATCTAGT' ]

# Diccionario que almacena la conversión de codon a aminoácido
CODE = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'}

def dna_aa(seq):
  # 1. Dividir la secuencia en secuencia
  lista_codon = []
  
  # Si la secuencia tiene 21 nt. "i" recorre los índices de 0 a 20, saltando 3 espacios. 
  # Es decir i toma los indices 0,3,6,9,12,15 y 18
  for i in range(0,len(seq),3):
    # El codon toma 3 caracteres a la vez
    codon = seq[i:i+3]
    #almacenamos el codon en una lista
    lista_codon.append(codon)

  # 2. Hacer la conversión de codon a aminoácido usando el diccionario
  lista_amino = []
  #por cada codon
  for elemento in lista_codon:
  
    # Uso el codon para obtener su aminoácido correspondiente. Uso el diccionario
    amino = CODE[elemento]
    # Almaceno el aminoácido en una lista
    lista_amino.append(amino)
  
  # 3. La lista de aminoácidos lo convierto en un solo string
  seq_amino = ''.join(lista_amino)
  return seq_amino
# Escribimos en el archivo
nombre =  '/content/drive/MyDrive/COLAB/database3.txt'
archivo = open(nombre, 'w')

for i in range(len(secuencias)):
  id = str(nombres[i])
  dna = str(secuencias[i])
  aa = str(dna_aa(secuencias[i]))
  archivo.write( id +',' + dna + ',' + aa +'\n')
 
# Cerramos el archivo
archivo.close()


    
````

````python
# Archivo secuencias.fasta
dir =  '/content/drive/MyDrive/COLAB/fasta_read.fasta'
# Abrimos el archivo para añadir texto al final del archivo
archivo = open(dir,'r')
data = archivo.read()
def fastaread(data):

  # data = '>A22886.1 CCV-C54 spike gene\nTTGCTCATTAGAAACAATGGAAAACTACTAAACTTCGGTAATCACTTGGTTAATGTGCCATGATTGTGCT\nACGTCCACTAAATTCAAAACTAATA\n\n>A22884.1 CCVI spike gene\nTTGCTCATTAGAAACAATGGTAAACTACTAAACTTTGGTAATCACTTGGTTAATGTGCCATGATTGTGCT\nACTGAATTCAAAATGTTAA\n\n>A22882.1 CCV6 spike gene\nGTAAGTTGCTCATTAGAAACAATGGAAAACTACTAAACTTTGGTAATCACTTGGTTAATGTGCCATGATT\nCAACTATTAGCATTAATAAA\n'

  # Eliminamos los \n al final del string con 
  lista = data.rstrip('\n')
 
  # Dividimos usando el metodo split()
  lista = lista.split('\n\n')

  # lista = ['>A22886.1 CCV-C54 spike gene\nTTGCTCATTAGAAACAATGGAAAACTACTAAACTTCGGTAATCACTTGGTTAATGTGCCATGATTGTGCT\nACGTCCACTAAATTCAAAACTAATA', 
  #          '>A22884.1 CCVI spike gene\nTTGCTCATTAGAAACAATGGTAAACTACTAAACTTTGGTAATCACTTGGTTAATGTGCCATGATTGTGCT\nACTGAATTCAAAATGTTAA', 
  #          '>A22882.1 CCV6 spike gene\nGTAAGTTGCTCATTAGAAACAATGGAAAACTACTAAACTTTGGTAATCACTTGGTTAATGTGCCATGATT\nCAACTATTAGCATTAATAAA']
  nombre_lista = []
  secuencia_lista = []


    # elemento_como_lista = ['>A22886.1 CCV-C54 spike gene', 'TTGCTCATTAGAAACAATGGAAAACTACTAAACTTCGGTAATCACTTGGTTAATGTGCCATGATTGTGCT', 'ACGTCCACTAAATTCAAAACTAATA']
  for elemento in lista:
        elemento_como_lista = elemento.split('\n')
    # name = ['A22886.1 CCV-C54 spike gene']
        name = elemento_como_lista[0][1:]
        lista_secuencia = elemento_como_lista[1:]
        separator = ''
        sequence = separator.join(lista_secuencia)
    
    # lista_secuencia = ['TTGCTCATTAGAAACAATGGAAAACTACTAAACTTCGGTAATCACTTGGTTAATGTGCCATGATTGTGCT', 'ACGTCCACTAAATTCAAAACTAATA']
        nombre_lista.append(name) 
        secuencia_lista.append(sequence)

    # sequence = ['TTGCTCATTAGAAACAATGGAAAACTACTAAACTTCGGTAATCACTTGGTTAATGTGCCATGATTGTGCTACGTCCACTAAATTCAAAACTAATA']
  return nombre_lista,secuencia_lista
    

  
````
