**ACCEDER A UN STRING**

Un string es una secuencia de caracteres. Podemos acceder a los caracteres uno por uno si usamos corchetes `[]`. El valor dentro de los corchetes se conoce como índice. El índice nos indica cual caracter en la secuencia queremos acceder. 

```python
verdura = 'espinaca'

# ¿Qué número debemos colocar si queremos acceder a la primera letra?
# Probemos con 1 
verdura[1]
```

* Para muchos, la primera letra de espicana es `e` y no `s`. Pero en python, es diferente. Para acceder la primera letra de un string debemos colocar el índice `0`

```python
verdura[0]
```


De esta manera `e` corresponde al índice `[0]`, s al índice `1`, `p` al índice `2` y así sucesivamente.

>> **Realiza 5:** Usando el bucle for y la expresión if, genere un script que cuente la vocal `'a'` de la palabra `'radiografía'`

```python
palabra = 'radiografia'

#iniciamos en cero la variable 'contar'
contar = 0

# Por cada letra en la palabra radiografia
for letra in palabra:
    #hacer el condicional, si la letra es la vocal 'a'
    if letra == 'a':
        # hacemos un incremento cada vez que encontramos la vocal 'a'
        contar = contar + 1

# Imprime la cantidad de vocales 'a'
print(contar)
```

```python
# Contar las t  del siguiente primer
primer = 'actgatcgatcgtactacgatcgtatgtacgtaatcgta'
#iniciamos en cero la variable 'contar'
contar = 0

# Por cada letra en la palabra radiografia
for letra in primer:
    #hacer el condicional, si la letra es la vocal 'a'
    if letra == 't':
        # hacemos un incremento cada vez que encontramos la vocal 'a'
        contar = contar + 1

# Imprime la cantidad de vocales 'a'
print(contar)
```

>> **Realiza 6:** Convierta el ejercicio anterior en una función y calcule la cantidad de vocales 'a' de las siguientes palabras:
* electroencefalograma 
* electrocardiograma
* arteriosclerosis

```python
# variable de entrada : palabra
# variable de salida : contar

def contar_a(palabra):
  #iniciamos en cero la variable 'contar'
  contar = 0

  # Por cada letra en la palrba radiografia
  for letra in palabra:
      #hacer el condicional, si la letra es la vocal 'a'
      if letra == 'a':
          # hacemos un incremento cada vez que encontramos la vocal 'a'
          contar = contar + 1

  # Asignamos la variable de salida con RETURN
  return(contar)

print(contar_a('electroencefalograma'))
print(contar_a('electrocardiograma'))
print(contar_a('arteriosclerosis'))
```

```python
#Vamos a utilizar append para escribir una función, 
#powers_of_2 (n), que crea una lista de 
#potencias de 2 para todos los números enteros menores que n
def pw_2(n):
  powers=[]
  for i in range(n):
    powers.append(2**i)
  return powers

print(pw_2(3)) # range 0-->3  2**i --> i rango de 0 a 3
print(pw_2(5)) # range 0-->5  2**i --> i rango de 0 a 5
```
