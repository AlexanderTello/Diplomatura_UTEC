## Algebra de Matrices

$B=A*X$
\
$\begin{alignedat}{2}
-6&=3x+2y+z-w\\
17&=-2x-y+2z+w\\
-14&=5x-2y-2z-w\\
12&= x+y+2z+5w\\
\end{alignedat}$


${y} =A(\mathbf  {x} )={\begin{pmatrix}a_{11}x_{1}+\cdots +a_{1n}x_{n}\\a_{21}x_{1}+\cdots +a_{2n}x_{n}\\\vdots \\a_{m1}x_{1}+\cdots +a_{mn}x_{n}\end{pmatrix}}$

$\begin{matrix}a_{11}x_{1}+\cdots +a_{1n}x_{n}=b_{1}\\a_{21}x_{1}+\cdots +a_{2n}x_{n}=b_{2}\\\vdots \\a_{m1}x_{1}+\cdots +a_{mn}x_{n}=b_{m}\end{matrix}$





${x} =\begin{pmatrix}x_{1}\\x_{2}\\\vdots \\x_{n}\end{pmatrix}$
