#### crear dataframe desde un dicionario

```python
MD={'Edad': [35, 29, 52, 40, 62, 78, 59],
    'diagnostico': ['covid-19','neumonia','asma','cop', 'adrs','gripe','sano'],
    'departamento': ['lima', 'Callao', 'Cuzco', 'Huanuco', 'Loreto','Trujillo', 'Junin'],
    'altura': [171, 171, 163, 163, 163, 172, 155],
    'peso': [70, 83, 85, 73, 85, 95, 62]
    }

mi=['a01', 'b03', 'c04', 'd06', 'e05','w50', 'z99']
df = pd.DataFrame(MD,index=mi)
 
df
```

#### operaciones logicas

```python
print("\n -- loc -- \n")
print(df.loc[df['Edad'] < 50, ['departamento', 'diagnostico']])
 
```

```python
print("\n -- iloc -- \n")
print(df.iloc[(df['Edad'] < 40).values, [1, 3]])
```

#### selecion de datos con multiple condiciones

```python
df[(df['altura']< 165)|(df['Edad']>50)]# | es OR  &
```

```python
df[(df['altura']< 165)&(df['Edad']>40)]# es|or  es & and
```

#### agregar columna

```python
# 铆ndice de masa corporal (IMC), que se calcula dividiendo los kilogramos 
#de peso por el cuadrado de la estatura en metros (IMC = peso [kg]/ estatura^2[m2])

peso=df.peso.values
estatura=df.altura.values/100

IMC= peso/(estatura**2)
df['IMC']=np.round(IMC, 1)
df
```

#### sort de columnas

```python
df['IMC']=df['IMC'].astype(int)
cols=[ 'diagnostico', 'departamento', 'altura', 'peso','Edad', 'IMC']
ldf=df[cols].sort_values(['altura','peso','Edad'], ascending = (True,False,False,))
ldf
```

#### cambio de nombre de columnas

```python
ldf=ldf.rename(columns={'diagnostico':'Diagnostico', 'departamento':'Region','altura': 'Altura','peso':'Peso', 'Edad':'edad', 'IMC':'IMC'})
ldf
```
